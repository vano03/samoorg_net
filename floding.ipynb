```python
import random
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict

class Node:
    def __init__(self, x, y, frames):
        self.x = x
        self.y = y
        self.frames = frames
        self.energy = 100  # Начальное количество энергии

    def __repr__(self):
        return f"Узел({self.x}, {self.y}, кадры={self.frames}, энергия={self.energy})"

def is_connected(node1, node2, radius):
    distance = np.sqrt((node2.x - node1.x) ** 2 + (node2.y - node1.y) ** 2)
    return distance <= radius

def transfer_frames(nodes, radius, round_stat):
    gateway_frames = set()

    for current_node in nodes:
        for other_node in nodes:
            if current_node != other_node and is_connected(current_node, other_node, radius):
                unique_frames = [frame for frame in current_node.frames if frame not in other_node.frames]
                # Проверка, чтобы все уникальные кадры отправлялись шлюзу
                if unique_frames:
                    transferred_frames = unique_frames[:2]  # Передаем только первые 2 уникальных кадра
                    for frame in transferred_frames:
                        other_node.frames.append(frame)
                        gateway_frames.add(frame)
                        round_stat[(current_node, other_node)].append(frame)

    return gateway_frames

def generate_nodes(num_nodes):
    nodes = []
    for _ in range(num_nodes):
        x = random.randint(0, 300)
        y = random.randint(0, 300)
        frames = [random.randint(1, 100) for _ in range(random.randint(1, 5))]  # Генерация случайного количества кадров
        nodes.append(Node(x, y, frames))
    return nodes

def plot_graph(nodes, gateway_frames, radius, round_num):
    plt.figure(figsize=(8, 8))
    for node in nodes:
        plt.scatter(node.x, node.y, c='yellow' if node.energy > 0 else 'gray', s=100)
        plt.text(node.x, node.y, f' {node.frames}', fontsize=9, verticalalignment='bottom', horizontalalignment='right')

    # Отображаем шлюз
    plt.scatter(0, 0, c='red', s=100)
    plt.text(10, 10, 'ШЛЮЗ', color='black')

    # Соединяем узлы
    for i in range(len(nodes)):
        for j in range(i + 1, len(nodes)):
            if is_connected(nodes[i], nodes[j], radius):
                plt.plot([nodes[i].x, nodes[j].x], [nodes[i].y, nodes[j].y], color='black', linewidth=0.5)

    plt.xlim(-20, 320)
    plt.ylim(-20, 320)
    plt.grid()
    plt.title(f'Раунд {round_num}: Узлы и Шлюз (Переданные кадры: {gateway_frames})')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

# Главная часть программы
num_nodes = 10
radius = 50
nodes = generate_nodes(num_nodes)  # Генерируем узлы один раз

total_stat = defaultdict(list)
all_frames_sent = set()

# Собираем все уникальные кадры
for node in nodes:
    all_frames_sent.update(node.frames)

round_num = 1
while True:
    print(f"=== Раунд {round_num} ===")

    # Сбор статистики о пакетах за этот раунд
    round_stat = defaultdict(list)
    gateway_frames = transfer_frames(nodes, radius, round_stat)

    # Сбор итогов по переданным кадрам
    transferred_frames = len(gateway_frames)
    print(f"Переданные кадры в этом раунде: {transferred_frames}")
    for (sender, receiver), frames in round_stat.items():
        print(f"{sender} отправил кадры к {receiver}: {frames}")

    total_stat.update(round_stat)  # Обновляем общую статистику
    plot_graph(nodes, gateway_frames, radius, round_num)

    # Проверка, все ли уникальные кадры были отправлены в шлюз
    if all_frames_sent.issubset(gateway_frames):
        print("Все уникальные кадры были переданы в шлюз.")
        break

    # Проверим, если не было ни одной передачи:
    if transferred_frames == 0:
        print("Не было передачи, прекращение раундов.")
        break

    round_num += 1

# Подведение итогов по всем раундам
print("=== Общая Статистика ===")
total_transfers = defaultdict(lambda: defaultdict(list))

for (sender, receiver), frames in total_stat.items():
    total_transfers[sender][receiver].extend(frames)

for sender, receivers in total_transfers.items():
    for receiver, frames in receivers.items():
        print(f"{sender} отправил всего {len(frames)} кадров к {receiver}: {frames}")
```
